### Optimal Number of Neurons in Hidden Layer

```python
def plot_decision_graphs(hidden_layer_sizes=[1, 2, 4, 8, 16, 32, 64, 128], learning_rate=1.2):
    plt.figure(figsize=(16, 32))

    for i, n_h in enumerate(hidden_layer_sizes, 1):
        plt.subplot(5, 2, i)

        parameters = build_model(X, Y, n_h, learning_rate, num_iterations=5000)
    
        plot_decision_boundary(lambda x: predict(parameters, x.T), X, Y)
    
        predictions = predict(parameters, X)
        accuracy = np.mean(predictions == Y) * 100
    
        plt.title(f"Accuracy for {n_h} hidden units: {accuracy:.6f}")

plot_decision_graphs()
```

It's visible that the model is overfitting when going from 64 hidden layers to 128. So let's try going from 64 hidden units until 100 and also reduce the learning rate to see if it improves the model performance.

```python
plot_decision_graphs(hidden_layer_sizes=[i for i in range(64, 100, 6)], learning_rate=3.25)
```

Let's finally use 70 hidden units in the second layer with a learning rate of 3.25
